## Demo: Forcasting Stock Performance in Our Portfolio

This program retrieves stock price data using the Alpaca API and then simulates future performance using Monte Carlo Simulation.

<font color='red'>Comment</font>

Here, we project a portfolio's price/cumulative return trajectories.

### Import Dependencies

<font color='red'>Comment</font>

See those imports below? Monte Carlo needs 2 things to work.
1. historical data
2. a way to run monte carlo!

```python
# Import libraries and dependencies
import os
import pandas as pd
import alpaca_trade_api as tradeapi
from MCForecastTools import MCSimulation
```

```python
# Load .env enviroment variables
from dotenv import load_dotenv
import pathlib

#load_dotenv()  # for mac
home = pathlib.Path.home() / ".env"  # for pc 1/2
load_dotenv(dotenv_path=home)  # for pc 2/2
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv("ALPACA_API_KEY")
alpaca_secret_key = os.getenv("ALPACA_SECRET_KEY")

api = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version = "v2"
)
```

### Get Past 3 Year's Worth of Microsoft and Coca-Cola Stock Price Data via Alpaca API Call

```python
# Set timeframe to '1D'
timeframe = "1D"

# Set start and end datetimes between now and 3 years ago.
start_date = pd.Timestamp("2018-01-01", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2021-01-01", tz="America/New_York").isoformat()

# Set the ticker information
tickers = ["MSFT","KO"]

# Get 3 year's worth of historical price data for Microsoft and Coca-Cola
df_ticker = api.get_barset(
    tickers,
    timeframe,
    start=start_date,
    end=end_date,
    limit=1000,
).df

# Display sample data
df_ticker.head()
```

### Simulate five year portfolio growth using Monte Carlo simulation

```python
# Configuring a Monte Carlo simulation to forecast five years cumulative returns
MC_fiveyear = MCSimulation(
    portfolio_data = df_ticker,
    weights = [.60,.40],
    num_simulation = 500,
    num_trading_days = 252*5
)
```

```python
# Printing the simulation input data
MC_fiveyear.portfolio_data.head()
```

```python
# Running a Monte Carlo simulation to forecast five years cumulative returns
MC_fiveyear.calc_cumulative_return()
```

```python
# Plot simulation outcomes
line_plot = MC_fiveyear.plot_simulation()

# Save the plot for future usage
line_plot.get_figure().savefig("MC_fiveyear_sim_plot.png", bbox_inches="tight")
```

<font color='red'>Comment</font>

Pretty, but not very informative. We can't see the distribution. This IS a distribution, on it's side, but it's too hard to see. A better visual is immediately below:

```python
# Plot probability distribution and confidence intervals
dist_plot = MC_fiveyear.plot_distribution()

# Save the plot for future usage
dist_plot.get_figure().savefig('MC_fiveyear_dist_plot.png',bbox_inches='tight')
```

<font color='red'>Comment</font>

How often is the stock likely to fall within this range of cumulative returns?

<font color='red'>Comment</font>

Confidence Intervals!

> A...confidence interval gives a range of plausible values for the parameter of interest.

~https://www.khanacademy.org/math/ap-statistics/xfb5d8e68:inference-categorical-proportions/introduction-confidence-intervals/a/interpreting-confidence-levels-and-confidence-intervals

At a certain confidence level, such as 95%

As confidence goes up, the range gets wider. = I am 99.999999% confidence I will live between 1 day and 100 years. I am 0.000001% confident I will live 233,345 minutes.

95% chance observations fall within my range.

95% of coin flips fall between this range.

95% of returns fall between this range.

```python
# Fetch summary statistics from the Monte Carlo simulation results
tbl = MC_fiveyear.summarize_cumulative_return()

# Print summary statistics
print(tbl)
```

```python
tbl[8]
```

<font color='red'>Comment</font>

`ci_lower = round(tbl[8]*10000,2)` is kind of hard to interpret. Without the comment and print, you'd not really know what was happening. This is where carving out `initial_investment = 10000` could have added clarity.

```python
# Use the lower and upper `95%` confidence intervals to calculate the range of the possible outcomes of our $10,000 investments in Coca-Cola and Microsoft stocks
ci_lower = round(tbl[8]*10000,2)
ci_upper = round(tbl[9]*10000,2)

# Print results
print(f"There is a 95% chance that an initial investment of $10,000 in the portfolio"
      f" over the next 5 years will end within in the range of"
      f" ${ci_lower} and ${ci_upper}")
```

```python
initial_investment = 10000
# using actual terms like this, makes the formula much more informational
lower_95 = tbl[8]

ci_lower = round(lower_95 * initial_investment, 2)
# print("Lower estimate: $" + ci_lower)  # no. cannot concat Int to Str!
print("Lower estimate: $", ci_lower)  # yes, but notice the extra space
print("Lower estimate: $" + str(ci_lower))  # yes
print(f"Lower estimate: ${ci_lower}")  # yes, yay for f strings!7
```
